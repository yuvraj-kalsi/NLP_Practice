```python
import re
def is_abbreviation(word):
    pattern = r"^[A-Z]+$"
    return bool(re.match(pattern, word))
        
```

```python
# First round of cleaning
import re
import string

def clean_text_round1(text):

    # for i in text.split(' '):
    #     if len(i) > 1:
    #         if is_abbreviation(i) == True:
    #             print(i)
    #             if valid(i) == True:
    #                 print("Valid")

    text = text.lower()                  # lower-case
    text = re.sub('\[.*?\]','', text)    # for square-brackets
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)    # for punctuations
    # text = re.sub(r'[^\w\s]', '', text) 
    text = re.sub('\w*\d\w*', '', text)  # for string with numbers
    return text


round1 = lambda x : clean_text_round1(x)
```

### Round-3 of Cleaning

```python
import pandas as pd
import re

def check(t):
    file = pd.read_csv("Abbreviations and Slang.csv")

    if t in file['Abbreviations'].values:
        result = file.loc[file['Abbreviations'] == t,'Text'].values[0]
        return result
    else: 
        return False
```

```python
import enchant
def valid(t):

    invalids = []
    dictionary = enchant.Dict("en_US")
    
    for i in t.split(' '):
        
        if len(i) != 0:
            if dictionary.check(i):
                continue
            
            else:
                invalids.append(i)

    return invalids
```

```python
import enchant

def clean_text_round3(text):
        
    # 1 Removing Dash (like hyphens)
    # e.g. i– i– i you–
    text = re.sub('(?<=[a-z])–', '', text)
    text = re.sub('[0-9]–', '', text)
    text = re.sub('–','',text)

    # 2 Removing repeated words
    # e.g. i i i 
    text = re.sub(r'\b(\w+)(\W+\1\b)+', r'\1', text)
    # e.g. thatthat 
    text = re.sub(r'\b(\w+)(\1\b)+', r'\1', text)
       
    # 3 Removing emojis (e.g. ♪ ♪)
    emoj = re.compile("["
        u"\U0001F600-\U0001F64F"               
        u"\U0001F300-\U0001F5FF"               
        u"\U0001F680-\U0001F6FF"               
        u"\U0001F1E0-\U0001F1FF"               
        u"\U00002500-\U00002BEF"               
        u"\U00002702-\U000027B0"
        u"\U00002702-\U000027B0"
        u"\U000024C2-\U0001F251"
        u"\U0001f926-\U0001f937"
        u"\U00010000-\U0010ffff"
        u"\u2640-\u2642" 
        u"\u2600-\u2B55"
        u"\u200d"
        u"\u23cf"
        u"\u23e9"
        u"\u231a"
        u"\ufe0f"  
        u"\u3030"
                      "]+", re.UNICODE)

    text = re.sub(emoj, '', text)


    # 4 Replacing contractions with its valid words
    # (e.g. n't -> not, 'll -> will, 're -> are)   
    a = {"s" : " is", "m" : " am", "nt" : " not", "im" : "i am", "ll" : " will",
        "ve" : " have", "re" : " are", "youre" : " you are",
        "isnt" : " is not", "arent" : " are not",
        "dont" : " do not", "doesnt" : " does not", "didnt" : " did not", 
        "havent" : " have not", "hasnt" : " has not", "hadnt" : " had not",
        "wasnt" : " was not", "wont" : " will not", "wouldnt" : "would not", 
        "cant" : " cannot", "couldnt " : "could not",
        "theres" : "there is", "thats" : "that is", "weve" : "we have",
        "ive" : "i have", "youve" : "you have", "theyve" : "they have", 
        "wassup" : "whatsup", "theyre" : "they are", "theyll" : "they will",
        "youll" : "you will", "well" : "we will", "ill" : "i will", "shouldve" : "should have",
        "whos" : "who is"}

    for key in a:
        for i in text.split(' '):
            if i == key:
                regex = r'\b' + re.escape(key)           # will not search for 'im' in 'time'
                text = re.sub(regex, a[key], text)

    # 5 Replacing short forms with their full forms using data-set
    # (e.g. omg -> oh my god)   


    # for i in text.split(' '):
    #      if check(i) != False:
    #         regex = r'\b' + re.escape(i)     
    #         text = re.sub(regex, check(i), text)
    #         # print(i, " ", check(i))


    return text

    
round3 = lambda x : clean_text_round3(x)
```

### Document-Term Matrix

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(stop_words = 'english', ngram_range = (1,2), min_df = 2, max_df = 0.50)

# ngram_range = (1,2) -> it will consider unigram as well as bi-gram
# min_df = 2 -> remove words which are occuring in less than 2 documents
# max_df = 0.50 -> remove word which are occuring in 50% of the documents

data_cv = cv.fit_transform(data_clean.transcript)
data_dtm = pd.DataFrame(data_cv.toarray(), columns=cv.get_feature_names_out())
data_dtm.index = data_clean.index
data_dtm
```
