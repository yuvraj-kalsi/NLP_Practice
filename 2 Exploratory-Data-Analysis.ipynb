## Most Common Words

```python
import pandas as pd

data = pd.read_pickle('new pickle/dtm.pkl')
data = data.transpose()
data.head()
```

```python
top_dict = {}
for c in data.columns:
    top = data[c].sort_values(ascending = False).head(30)
    top_dict[c] = list(zip(top.index, top.values))

top_dict
```

```python
for comedians, top_words in top_dict.items():
    print(comedians)
    print(', '.join([word for word, count in top_words[0:14]]))
    print('---')
```

```python
from collections import Counter

words = []
for comedians in data.columns:
    top = [word for word, count in top_dict[comedians]]
    for t in top:
        words.append(t)


words
```

```python
Counter(words).most_common()
```

```python
# 8 because -> half of the comedians
add_stop_words = [word for word, count in Counter(words).most_common() if count > 8]
add_stop_words
```

```python
import pandas as pd
from sklearn.feature_extraction import text
from sklearn.feature_extraction.text import CountVectorizer


data_clean = pd.read_pickle('new pickle/data_clean.pkl')

stop_words1 = text.ENGLISH_STOP_WORDS.union(add_stop_words)
stop_words = list(stop_words1)

cv = CountVectorizer(stop_words = stop_words)
data_cv = cv.fit_transform(data_clean.transcript)
data_stop = pd.DataFrame(data_cv.toarray(), columns = cv.get_feature_names_out())
data_stop.index = data_clean.index

print(data_stop)

import pickle
pickle.dump(cv, open('cv_stop.pkl', 'wb'))
data_stop.to_pickle('dtm_stop.pkl')
```

```python
from wordcloud import WordCloud

wc = WordCloud(stopwords = stop_words, background_color = "white", colormap = "Dark2",
                max_font_size = 150, random_state = 42)
```

```python
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = [16, 6]

full_names = ['Ali Wong', 'Anthony Jeselnik', 'Bill Burr', 'Bo Burnham', 'Dave Chappelle', 'Gabriel Iglesias', 
            'Hasan Minhaj', 'Jack Whitehall', 'Jim Jefferies', 'Joe Rogan', 'John Mulaney', 'Louis C.K.', 'Mike Birbiglia', 
            'Ricky Gervais', 'Trevor Noah', 'Vir Das']

for index, comedian in enumerate(data.columns):

    wc.generate(data_clean.transcript[comedian])

    plt.subplot(4, 4, index + 1)

    plt.imshow(wc, interpolation = "bilinear")

    plt.axis("off")

    plt.title(full_names[index])

plt.tight_layout()
plt.show()
```

## Number of Words

```python
import numpy
unique_list = []
for comedian in data.columns:
    uniques = data[comedian].to_numpy().nonzero()[0].size
    unique_list.append(uniques)

print(unique_list)
data_words = pd.DataFrame(list(zip(full_names, unique_list)), columns = ['comedian', 'unique_words'])
data_unique_sort = data_words.sort_values(by = 'unique_words')
data_unique_sort
```

```python
total_list = []
for comedian in data.columns:
    totals = sum(data[comedian])
    total_list.append(totals)


run_times = [60, 59, 80, 60, 67, 110, 73, 58, 77, 63, 62, 58, 76, 79, 60, 50]

data_words['total_words'] = total_list
data_words['run_times'] = run_times
data_words['words_per_minute'] = data_words['total_words'] / data_words['run_times']

data_wpm_sort = data_words.sort_values(by = 'words_per_minute')
data_wpm_sort
```

```python
import numpy as np

y_pos = np.arange(len(data_words))

plt.subplot(1, 2, 1)
plt.barh(y_pos, data_unique_sort.unique_words, align = 'center')
plt.yticks(y_pos, data_unique_sort.comedian)
plt.title("Number of Unique Words", fontsize = 30)

plt.subplot(1, 2, 2)
plt.barh(y_pos, data_wpm_sort.words_per_minute, align = 'center')
plt.yticks(y_pos, data_wpm_sort.comedian)
plt.title("Number of Words per Minute", fontsize = 30)

plt.tight_layout()
plt.show()
```
